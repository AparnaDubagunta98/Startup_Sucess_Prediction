```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold
from keras.wrappers.scikit_learn import KerasClassifier
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.layers import BatchNormalization #for batch normalization
from keras.layers import Dropout            #for random dropout
from tensorflow.keras.optimizers import Adam  #for adam optimizer
from keras import regularizers      #for l2 regularization
from sklearn.model_selection import RandomizedSearchCV
```

```python
df = pd.read_csv("success_encoded_data.csv")

```

```python
success_df = df[df['target_variable'] == 1]
failure_df = df[df['target_variable'] == 0].sample(n = success_df.shape[0], replace = False)
balanced_df = pd.concat([success_df,failure_df])
balanced_df = balanced_df.sample(frac=1).reset_index(drop=True)
df = balanced_df
```

```python
df.drop('company_name', axis=1, inplace=True)
df.drop('company_code',axis=1,inplace=True)
y = df['target_variable']
df.drop('target_variable', axis=1, inplace=True)
df.drop(['acquired','closed','ipo','operating'],axis=1,inplace=True)
```

```python
X = df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)

# preds = [i[0] for i in preds]
# final_predictions_FNN.extend(preds)
```

```python
def nn_model(nn1,lr,opt):
    input_dim = 733
    output_dim = 1
    model = Sequential()
    model.add(Dense(nn1, input_dim=input_dim, activation='tanh'))
    # model.add(Dense(5, input_dim=input_dim, activation='tanh'))
    model.add(Dense(1,activation='sigmoid'))
    # optimizer = tf.keras.optimizers.Adam(lr)
    model.compile(loss = 'binary_crossentropy',optimizer=opt,metrics=['accuracy'])
    return model
```

```python
lr=[1e-2, 1e-3, 1e-4]
nn1=[50,100,150,200]
optimizer = ['RMSprop','Adam']
param_grid = dict(lr=lr, nn1=nn1, opt = optimizer)
model = KerasClassifier(build_fn=nn_model, epochs=6, batch_size=20, verbose=1)
grid = RandomizedSearchCV(estimator=model, cv=3, param_distributions=param_grid, verbose=20,  n_iter=10, n_jobs=1)
grid_result = grid.fit(X_train, y_train,epochs=100,verbose=1)
```

```python
model = nn_model(733,0.01)
model.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
preds = model.predict(X_test)
```

```python
preds = [i[0] for i in preds]
```
